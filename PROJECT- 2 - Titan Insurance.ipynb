The Titan Insurance Company has just installed a new incentive payment scheme for its lift policy sales force. It wants to have an early view of the success or failure of the new scheme. Indications are that the sales force is selling more policies, but sales always vary in an unpredictable pattern from month to month and it is not clear that the scheme has made a significant difference.

For example, suppose salesperson X has, in the month, sold seven policies for which the sums assured are £1000, £2500, £3000, £5000, £10000, £35000. The scheme has now been in operation for four months. It has settled down after fluctuations in the first two months due to the changeover.

To test the effectiveness of the scheme, Titan have taken a random sample of 30 salespeople measured their output in the penultimate month prior to changeover and then measured it in the fourth month after the changeover (they have deliberately chosen months not too close to the changeover). ## Question 2

* Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
# Null hypothesis (H0): Data has not changed significently
# Alternate hypothesis (H1): Data has changed significently

from scipy.stats import ttest_ind
from statsmodels.stats.power import ttest_power

t_statistic, p_value = ttest_ind(ti_df['Old Scheme (in thousands)'], ti_df['New Scheme (in thousands)'])
# t_statistic, p_value = ttest_ind(ti_df['New Scheme (in thousands)'], ti_df['Old Scheme (in thousands)'])

# Print the output
print('P_value is ', p_value)

if(p_value < 0.05):
    print('As per the p_value the null-hypothesis is rejected at 5% significance. \nSignificant change is observed between the old and new scheme.')
else:
    print('As per the p_value the null-hypothesis is accepted at 5% significance. \nSignificant change is not observed between the old and new scheme.')
```

## Question 4

* Suppose it has been calculated that in order for Titan to break even, the average output must increase by £5000 in the scheme compared to the old scheme. If this figure is alternative hypothesis, what is:

 a) The probability of a type 1 error? (2.5 points)

 b) What is the p- value of the hypothesis test if we test for a difference of $5000? (10 points)

 c) Power of the test (5 points)



```python
# Create a new column with difference between new scheme and old scheme

ti_df['New to old delta'] = (ti_df['New Scheme (in thousands)'] - ti_df['Old Scheme (in thousands)'])

ti_df.head()
ti_df.describe().T
```

```python
# One Sample T-test

# Type 1 error (false positive): Rejected the null hypothesis, but it was correct
# Type 2 error (false negative): Retained the null hypothesis, but alternate hypothesis was correct

# Null Hypothesis (H0): The average increase in output is not more than 5000
# Alternate Hypothesis (H1): Average output increase is more than £5000 in the scheme compared to the old scheme

from scipy.stats import ttest_1samp

t_statistic, p_value = ttest_1samp(ti_df['New to old delta'], 5)

print("t_statistic, p_value: ", t_statistic, p_value)
```

###### 4) a. The probability of a type 1 error? (2.5 points)

```python
print("The probability of getting type 1 error is given by p-value, which is: ", p_value)
```

###### 4) b. What is the p- value of the hypothesis test if we test for a difference of $5000? (10 points)

```python
print('The p-value generated from 1 sample t-test is: ', p_value)
```

###### 4) c. Power of the test (5 points)

```python
import statsmodels.stats.power as smp

# z_stat = 5/np.std(ti_df['New Scheme (in thousands)'])
# nobs = len(np.array(ti_df['New Scheme (in thousands)'])) - 1

z_stat = 5/np.std(ti_df['New to old delta'])
nobs = len(np.array(ti_df['New to old delta'])) - 1

print("z_stat (effect size): ", z_stat)
print("nobs: ", nobs)

power = smp.ttest_power(effect_size=z_stat, nobs=nobs, alpha=0.05, alternative='larger')

print("Power of the t_test is", power)
```
