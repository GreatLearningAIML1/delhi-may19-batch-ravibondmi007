### Write a function that accepts two numbers as input and prints them - Pass the Sepal length and sepal width of 5th row and print the output

```python
def printVal(x, position):
    try:
        int(x)
        float(x)
        print('Input value at %s input position is:'%(position))
        print(x)
    except:
        print('Input value at %s position is not of number type. Please validate and retry.'%(position))

def printNums(a,b):
    printVal(a, 'first')
    printVal(b, 'second')
    return

printNums(Iris.iloc[4,0], Iris.iloc[4,1])
```
